```python
import numpy as np
import random
from itertools import permutations
from numpy.random import choice

lim = 5
```

```python
def count_sublist(lst, sublst):
    count = 0
    sublst_length = len(sublst)
    
    for i in range(len(lst) - sublst_length + 1):
        if lst[i:i + sublst_length] == sublst:
            count += 1
    
    return count
```

```python
def pprint(mat):
    for m in mat:
        print(m)
```

```python
source = [random.randint(0,lim-1) for i in range(100)]
```

```python
matrix = []
for i in range(lim):
    arr = []
    for j in range(lim):
        arr.append(count_sublist(source,[i,j]))
    matrix.append(arr)
pprint(matrix)
```

```python
for i in range(len(matrix)):
    s = sum(matrix[i])
    for j in range(len(matrix[i])):
        matrix[i][j]/=s

pprint(matrix)

for m in matrix:
    if sum(m) != 1:
        print("lolz")
```
